```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://arkhamdb.com/find?q=f%3Aguardian&decks=player'
html = requests.get(url)

soup = BeautifulSoup(html.content, 'html.parser')

results = soup.find(id='list')

titles = results.find_all('a', class_='card-tip')
```

```python
for title in titles:

    print(title.text)
```
